```python
def get_filename_from_ax(ax):
    title = ax.get_title().split(" ")
    s = ""
    for i,f in enumerate(title):
        if i <= len(title)-1 and i > 0:
            s+="_"+f
        else:
            s+=f
    if '-' in s:
        s = s.replace("-","_")
    if '=' in s:
        s = s.replace("=","_")
    return s
```
